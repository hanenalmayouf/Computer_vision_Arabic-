## Read and Show Images

To read an image, we need to use `Image.open()`, and to convert the image to an `np.ndarray`, we will use `np.asarray()`.

````python
import os
import requests
image_file_path = "horse.png"
# Download the image if it doesn't exist (e.g. in Colab)
if not os.path.exists(image_file_path):
    url = f"https://raw.githubusercontent.com/NaifMersal/cv-for-developers-ultralytics/main/labs/{image_file_path}"
    r = requests.get(url)
    with open(image_file_path, "wb") as f:
        f.write(r.content)
img = Image.open(image_file_path)
arr = np.asarray(img)
````

To show the image, we will have to convert it back to a `PIL.Image.Image` using `PIL.Image.fromarray()`.

````python
Image.fromarray(arr)
````
