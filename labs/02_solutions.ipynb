## 3. Object Counting

````python
from ultralytics import YOLO
from ultralytics.solutions import ObjectCounter

model = YOLO("weights/yolo26n.pt")
video_path = "https://ultralytics.com/videos/Cars.mp4"

counter = ObjectCounter(names=model.names, view_img=True, region=[(100, 300), (700, 300)])

results = model.track(source=video_path, show=False, stream=True, classes=[2,3,5,7])

for result in results:
    counter.start_counting(frame=result.plot(), boxes=result.boxes)

print("\nFinal Counts:")
for cls, count in counter.classwise_counts.items():
    if count > 0:
        print(f"  {cls}: {count}")
````

## 4. Heatmap Visualization

````python
from ultralytics import YOLO
from ultralytics.solutions import Heatmap

model = YOLO("weights/yolo26n.pt")
video_path = "https://ultralytics.com/videos/Cars.mp4"

heatmap = Heatmap(names=model.names, view_img=True)
results = model.predict(source=video_path, show=False, stream=True, classes=[2,3,5,7])

for result in results:
    heatmap.generate_heatmap(imgsz=(result.orig_img.shape[0], result.orig_img.shape[1]), boxes=result.boxes)

print("\nHeatmap complete!")
````
